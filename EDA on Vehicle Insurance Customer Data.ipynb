### checking and cleaning data quality

```python
# counting the null values of customer_data column wise
print(customer_data.isnull().sum())  
print("Total Count of null values of customer_data : ",customer_data.isnull().sum().sum())
```

```python
# dropping the null values for customer_id's only    
customer_data.dropna(axis='index',how='any',subset=['customer_id'],inplace=True) 
```

```python
customer_data
```

```python
# filling the categorical columns with their respective mode value

customer_data['Gender'].fillna(customer_data['Gender'].mode()[0],inplace=True)
customer_data['vehicle age'].fillna(customer_data['vehicle age'].mode()[0],inplace=True)
customer_data['vehicle damage'].fillna(customer_data['vehicle damage'].mode()[0],inplace=True)
```

```python
# filling the numeric columns with their respective mean value

customer_data['age'].fillna(customer_data['age'].mean(),inplace=True)
customer_data['driving licence present'].fillna(customer_data['driving licence present'].mean(),inplace=True)
customer_data['region code'].fillna(customer_data['region code'].mean(),inplace=True)
customer_data['previously insured'].fillna(customer_data['previously insured'].mean(),inplace=True)
```

### finding the outliers

```python
# calculating the iqr for whole datafram of customer_data
q1=customer_data.quantile(0.25)
q3=customer_data.quantile(0.75)
iqr=q3-q1
iqr
```

```python
# finding no of coutliers for each column
((customer_data > (q3+1.5*iqr)) | (customer_data < (q1-1.5*iqr))).sum() 
```

```python
print('Total Outliers :',((customer_data > (q3+1.5*iqr)) | (customer_data < (q1-1.5*iqr))).sum().sum()) 
```

## replacing outliers

```python
# finding the 25% percentile and 75% percentile for driving license present

q1= customer_data['driving licence present'].quantile(0.25)
q3= customer_data['driving licence present'].quantile(0.75)

IQR=q3-q1
IQR
```

```python
# copying customer_data into customer_data1 to see changes
customer_data1=customer_data.copy()

customer_data1['driving licence present']=np.where(
                                          customer_data1['driving licence present'] > (q3+1.5*IQR),
                                          
                                          customer_data1['driving licence present'].mean(),
                                             np.where(
                                                customer_data1['driving licence present'] < (q1-1.5*IQR),
                                                customer_data1['driving licence present'].mean(),
                                                 customer_data1['driving licence present']
                                                    )
                                                 )
```
